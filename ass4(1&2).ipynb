```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
df = pd.read_csv('Housing (1).csv')
print(df.dtypes)
summary = df.describe(percentiles=[.25, .5, .75, .90])
print(summary)
columns = df.columns.tolist()
print(columns)
sns.pairplot(df)
plt.show()
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
# Identify the dependent and independent variables
dependent_var = 'Price'
independent_vars = df.select_dtypes(include=['number']).columns.tolist()
independent_vars.remove(dependent_var)

print("Dependent Variable:", dependent_var)
print("Independent Variables:", independent_vars)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
X = np.random.rand(100, 1)
Y = 3 * X.squeeze() + np.random.randn(100)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, Y_train)
import scipy.stats as stats
intercept = model.intercept_
slope = model.coef_[0]
Y_pred = model.predict(X_test)
cdf_values = stats.norm.cdf(Y_pred, np.mean(Y_pred), np.std(Y_pred))
print(f"Intercept: {intercept}")
print(f"Slope: {slope}")
print(f"CDF of the predictions: {cdf_values}")
```
